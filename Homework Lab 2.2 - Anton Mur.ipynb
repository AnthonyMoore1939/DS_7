### Завдання 1

Для трьох введених натуральних чисел (користувач вводить числа через пробіл) знайдіть мінімальне та максимальне

```python
# Оскільки через min/max було б занадто просто - реалізувано циклами

numbers = input('Введіть три натуральні числа (через пробіл): ')

numbers_list = numbers.split()

max_list = float('-inf')
min_list = float('inf')

int_numbers = []

for i in range(len(numbers_list)):
    int_numbers.append(int(numbers_list[i]))
    
    if int_numbers[i] > max_list:
        max_list = int_numbers[i]
        
    i += 1
    
print('Максимальне число з введених: ', max_list)
    
for i in range(len(numbers_list)):
      
    if int_numbers[i] < min_list:
        min_list = int_numbers[i]
        
    i += 1
        
print('Мінімальне число з введених:', min_list)
```

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності

Послідовність чисел ганеруємо рандомно

```python
import random

#генеруэмо тисячу чисел в діапазоні від 0 до 100
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]

print(list_with_random_numbers)

#1. Розраховує мінімальне та максимальне значення // (!) без використання функцій min/max 

my_min = float('inf')
my_max = float('-inf')

for i in range(len(list_with_random_numbers)):
    if list_with_random_numbers[i] < my_min:
        my_min = list_with_random_numbers[i]
        
print('Найменше значення списку: ', my_min)

for i in range(len(list_with_random_numbers)):
    if list_with_random_numbers[i] > my_max:
        my_max = list_with_random_numbers[i]

print('Найбільше значення списку: ', my_max)
    
```

```python
# 2. Середнє значення послідовності

def average(list_with_random_numbers):
    s = 0
    for i in range(len(list_with_random_numbers)):
        s += list_with_random_numbers[i]
        i += 1
    return s // len(list_with_random_numbers)

average(list_with_random_numbers)
```

```python
# 3. Знайти медіану послідовності (функцією) без використання спеціалізованих бібліотек

list_sort = sorted(list_with_random_numbers)

def median(list_sort):
    
    # врахуємо, що кількість елементів списку може бути парним чи непарним.
    
    if len(list_sort) % 2 == 0:
        return (list_sort[len(list_sort) // 2] + list_sort[(len(list_sort) // 2) + 1]) // 2
    
    else:
        return list_sort[(len(list_sort) // 2) + 1]


print('Медіана послідовності: ', median(list_sort))
```

```python
# 4. Знайти частоту кожної цифри в послідовності / без Counter / 

def numbers(list_sort):
    
    numbers_dict = {}

    for i in range(len(list_sort)):
    
        if list_sort[i] in numbers_dict:
        
            numbers_dict[list_sort[i]] +=  1
             
        
        else:
        
            numbers_dict[list_sort[i]] = 1
    
        i += 1
        
    return numbers_dict

print('Частота повторення кожної цифри у послідовності: ', numbers(list_sort))
```

```python
# 5. Знайти середнє квадратичне відхилення

from math import sqrt

def my_st_dev(list_sort):
    
    sum_q = 0
    
    for i in range(len(list_sort)):
    
        sum_q += (list_sort[i] - average(list_with_random_numbers))**2
    
        i += 1
    
    st_dev = sqrt(sum_q / len(list_sort))
    
    return st_dev

print('Розрахункове стадартне відхилення = ', round(my_st_dev(list_sort), 2))


# для контролю - пошук відхилення спеціалізованою функцією. Можуть трохи не збігатися

import statistics

print('! Контроль = ', round(statistics.stdev(list_with_random_numbers), 2))
```
